```python
import pandas as pd

# 读取数据文件
data = pd.read_csv('data/air_data.csv')

# 打印数据的特征
print(data.columns)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import datetime
import sklearn.preprocessing
import sklearn.cluster
```

```python
data.head(5)
```

```python
# 查看数值特征的一些统计数据
data.describe().T
```

```python
# 去掉重复值
data=data.drop_duplicates()
```

```python
# 查空缺值
data.isnull().sum()
```

```python
load_time = datetime.datetime.strptime('2014/03/31', '%Y/%m/%d')
ffp_dates = [datetime.datetime.strptime(ffp_date, '%Y/%m/%d') for ffp_date in data['FFP_DATE']]
length_of_relationship = [(load_time - ffp_date).days for ffp_date in ffp_dates]
data['LEN_REL'] = length_of_relationship
```

```python
features = ['LEN_REL','FLIGHT_COUNT','avg_discount','SEG_KM_SUM','LAST_TO_END']
data = data[features]
features = ['L','F','C','M','R']
data.columns = features
```

```python
data.head()
```

```python
data.describe().T
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
data[features] = scaler.fit_transform(data[features])
```

```python
data.describe().T
```

```python
data.head()
```

```python
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=5).fit(data)
```

```python
# 获取每个类的中心（质心）
cluster_centers = kmeans.cluster_centers_
```

```python
cluster_centers
```

```python
import numpy as np
unique, counts = np.unique(kmeans.labels_, return_counts=True)
cluster_counts = dict(zip(unique, counts))
```

```python
for cluster_label, center in enumerate(cluster_centers):
    print(f"Cluster {cluster_label} Center: {center}")

for cluster_label, count in cluster_counts.items():
    print(f"Cluster {cluster_label} Count: {count}")
```

```python
 from sklearn.cluster import DBSCAN

# 创建DBSCAN对象并进行聚类
dbscan = DBSCAN(eps=0.5, min_samples=5)  # 根据需要设置eps和min_samples参数
clusters = dbscan.fit_predict(data)
```

```python
# 获取聚类种类和每个类的数量
unique_clusters, cluster_counts = np.unique(clusters, return_counts=True)

# 打印聚类种类和每个类的数量
for cluster, count in zip(unique_clusters, cluster_counts):
    print(f"Cluster {cluster}: {count} samples")
```
